```python
from src.main import main
from argparse import Namespace
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
test_cases = [
    Namespace(method="BFS", filename="data/cidade.map", xi=130, yi=1, xf=136, yf=10),
    Namespace(method="Astar", filename="data/cidade.map", xi=130, yi=1, xf=136, yf=10),
    Namespace(method="UCS", filename="data/cidade.map", xi=130, yi=1, xf=136, yf=10),
    Namespace(method="IDS", filename="data/cidade.map", xi=130, yi=1, xf=136, yf=10),
    Namespace(method="Greedy", filename="data/cidade.map", xi=130, yi=1, xf=136, yf=10),
]

test_cases2 = [
    Namespace(method="BFS", filename="data/floresta.map", xi=27, yi=11, xf=13, yf=13),
    Namespace(method="Astar", filename="data/floresta.map", xi=27, yi=11, xf=13, yf=13),
    Namespace(method="UCS", filename="data/floresta.map", xi=27, yi=11, xf=13, yf=13),
    Namespace(method="IDS", filename="data/floresta.map", xi=27, yi=11, xf=13, yf=13),
    Namespace(method="Greedy", filename="data/floresta.map", xi=27, yi=11, xf=13, yf=13),
]
```

```python
results1 = {}
results2 = {}

for i, args in enumerate(test_cases):
    print(f"Running test case {i+1}: {args}")
    start_time = time.time()
    cost, path = main(args)
    elapsed_time = time.time() - start_time
    results1[args.method] = {
        "cost": cost,
        "time_taken": elapsed_time,
        "path": path,
    }

for i, args in enumerate(test_cases2):
    print(f"Running test case {i+1}: {args}")
    start_time = time.time()
    cost, path = main(args)
    elapsed_time = time.time() - start_time 
    results2[args.method] = {
        "cost": cost,
        "time_taken": elapsed_time,
        "path": path,
    }
```

```python
data = []
for method in results1:
    data.append({"Method": method, "Cost": results1[method]["cost"], "Test Set": "Test Cases 1"})
for method in results2:
    data.append({"Method": method, "Cost": results2[method]["cost"], "Test Set": "Test Cases 2"})

df = pd.DataFrame(data)

# Criando subplots lado a lado
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1
axes[0].set_title("Resultados do Teste - Conjunto 1", fontsize=14)
for method, data in results1.items():
    path = data["path"]
    x_coords, y_coords = zip(*path)
    axes[0].plot(y_coords, x_coords, label=f"{method} (Custo: {data['cost']:.2f})")
axes[0].set_xlabel("Y")
axes[0].set_ylabel("X")
axes[0].legend()
axes[0].grid()

# Gráfico 2
palette = {
    "Test Cases 1": "#782580",
    "Test Cases 2": "#77DD77",
}

sns.barplot(data=df, x="Method", y="Cost", hue="Test Set", ax=axes[1], palette=palette)
axes[1].set_title("Comparação dos métodos quanto ao custo", fontsize=14)
axes[1].set_xlabel("Algoritmo")
axes[1].set_ylabel("Custo")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = []
for method, result in results1.items():
    if method == "IDS" or method == "BFS":
        continue
    data.append({"method": method, "cost": result["cost"], "time_taken": np.log(result["time_taken"])})
for method, result in results2.items():
    if method == "IDS" or method == "BFS":
        continue
    data.append({"method": method, "cost": result["cost"], "time_taken": np.log(result["time_taken"])})

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="time_taken", y="cost", hue="method", s=50, alpha=1)
plt.title("Custo vs tempo gasto")
plt.xlabel("(log scale) Tempo gasto (s)")
plt.ylabel("Custo")
plt.legend(title="Algoritmo")
plt.grid(True)
plt.show()
```
